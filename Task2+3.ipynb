```
df = pd.read_csv('/content/Task 3_Final Content Data set.csv')
df.head()
```

```
df = pd.DataFrame(df)
df.head()
```

```
# Group by 'Category' and calculate the sum of 'Score' for each category
category_scores = df.groupby('Category')['Score'].sum().reset_index()

# Sort the categories by score in descending order
category_scores
```

```
import matplotlib.pyplot as plt
import seaborn as sns

#pie chart

# Sort and select top 5 categories
top_categories = category_scores.sort_values(by='Score', ascending=False).head(5)

plt.figure(figsize=(4,6))
plt.pie(top_categories['Score'], labels = top_categories['Category'],autopct='%1.1f%%')
plt.title("Top 5 Categories")
plt.show()
```
