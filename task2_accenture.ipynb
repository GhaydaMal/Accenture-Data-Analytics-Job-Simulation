```
import pandas as pd
```

```
#load files
df1 = pd.read_csv('/content/Content.csv')
df2 = pd.read_csv('/content/ReactionTypes.csv')
df3 = pd.read_csv('/content/Reactions.csv')
#dataframe
df1 = pd.DataFrame(df1)
df2 = pd.DataFrame(df2)
df3 = pd.DataFrame(df3)
#visual
df3.head()
```

```
new_df = pd.merge(df1, df3, on='Content ID')
new_df.head()
```

```
df = pd.merge(new_df , df2, on='Reaction Type' )
df.head()
```

```
#to cvs format
df.to_csv('mergedData.csv')
```
